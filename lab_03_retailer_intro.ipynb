## Prepare Data

➡️ Assignment Tasks
- Create a "revenue" variable and make sure it is a numeric datatype
- Remove rows with empty (NULL) values
- Locate outliers in at least two of the variables.   Filter them out if they are mistakes; keep them if they are exceptional cases.

```python
#Creates a "revenue" variable and makes sure it is a numeric datatype
df['revenue'] = df['quantity'] * df['product_price']
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
```

```python
#Removes rows with empty (NULL) values
df.dropna(inplace=True)
```

```python
# Example for 'quantity' and 'product_price' columns
for column in ['quantity', 'product_price']:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    # Print some info about the outliers
    print(f"Outliers in {column}:")
    print(outliers[[column, 'revenue']])

    # Decision making and filtering (example logic)
    if column == 'quantity' and any(outliers['quantity'] < 0):
        df = df[(df['quantity'] >= lower_bound) & (df['quantity'] <= upper_bound)]
    elif column == 'product_price' and any(outliers['product_price'] > 1000):
        df = df[(df['product_price'] >= lower_bound) & (df['product_price'] <= upper_bound)]
    else:
        pass
```

## Summarize Data

➡️ Assignment Tasks
- Display the average price per product line
- Display the total revenue for all months

```python
#average price
average_price_per_product_line = df.groupby('product_line')['product_price'].mean()
print(average_price_per_product_line)
```

```python
#total revenue
total_revenue_per_month = df.groupby('order_year_month')['revenue'].sum()
print(total_revenue_per_month)
```
