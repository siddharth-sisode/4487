## Megatelco Data Dictionary

 DEMOGRAPHIC VARIABLES:
 - College - has the customer attended some college (one, zero)

PHONE VARIABLES:

ATTITUDINAL VARIABLES:

OTHER VARIABLES
 - Leave - Did this customer churn with the last contract expiration? (LEAVE, STAY)

Square bracket subsetting, or indexing, is a way to access specific rows and columns of a DataFrame. When you use square brackets, [], with a DataFrame you can pass in:

A single column name to select a single column (returns a Series)
A list of column names to select multiple columns (returns a DataFrame)
A boolean mask to select rows based on a condition
A slice to select a range of rows
For example, df['college'] selects the college column as a Series, while df[['college', 'income']] selects the college and income columns as a DataFrame.

## Data Visualization

 Assignment Tasks
- Add at least two charts to show a summary of any major trends

```python
import matplotlib.pyplot as plt  # Import pyplot

# Group data by operating system and leave status, then count occurrences
os_leave_counts = df_clean.groupby(['operating_system', 'leave'])['id'].count().unstack()

# Calculate percentages for stacking
os_leave_percentages = os_leave_counts.div(os_leave_counts.sum(axis=1), axis=0) * 100

# Create the stacked bar chart
ax = os_leave_percentages.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add labels and title using plt
plt.xlabel('Operating System')
plt.ylabel('Percentage of Customers')
plt.title('Operating System Distribution and Retention')
plt.legend(title='Leave Status', loc='upper right')

# Annotate bars with percentages
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height:.0f}%', (x + width / 2, y + height / 2), ha='center', va='center')
plt.show()
```

```python
#Display the proportion of customers with different satisfaction levels
import seaborn as sns

# Create the count plot
ax = sns.countplot(x='reported_satisfaction', hue='leave', data=df_clean)

# Add labels and title
plt.xlabel('Reported Satisfaction')
plt.ylabel('Number of Customers')
plt.title('Customer Satisfaction and Retention')

# Show the plot
plt.show()
```

```python


# Group data by data overage and calculate average call duration
overage_call_duration = df_clean.groupby('data_overage_mb')['average_call_duration'].mean()

# Create the line graph with adjusted aspect ratio
plt.figure(figsize=(12, 4))  # Adjust width and height as needed
plt.plot(overage_call_duration.index, overage_call_duration.values)

# Add labels and title
plt.xlabel('Data Overage (MB)')
plt.ylabel('Average Call Duration')
plt.title('Relationship between Data Overage and Average Call Duration')

# Show the plot
plt.show()
```
