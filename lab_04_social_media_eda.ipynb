## Prepare Data

‚û°Ô∏è Assignment Tasks
- Convert any numbers, such as quote_count, from strings into numbers
- Reduce the variables in your dataframe to only those that have useful data
- Optional: Locate outliers in any of the variables.   Filter them out if they are mistakes; keep them if they are exceptional cases.

```python
# Columns to convert (if they are read as strings)
string_number_cols = ['edit_history_tweet_ids', 'geo.place_id',
                      'attachments.media_keys', 'referenced_tweets']

# Convert to numeric, handling errors
for col in string_number_cols:
    if col in df.columns:  # Check if the column exists
        try:
            df[col] = pd.to_numeric(df[col], errors='coerce')
        except (ValueError, TypeError):  # Catch more specific errors
            print(f"Could not convert '{col}' to numeric. It might contain complex data.")
```

```python
useful_cols = ['author_id', 'created_at', 'text', 'lang',
               'public_metrics.quote_count', 'public_metrics.reply_count',
               'public_metrics.retweet_count', 'public_metrics.like_count']

# Select only the columns that actually exist in the DataFrame:
existing_cols = [col for col in useful_cols if col in df.columns]
df = df[existing_cols]
```

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Example: detecting and handling outliers in 'quote_count'
if 'quote_count' in df.columns:
    outliers_quote_count = detect_outliers_iqr(df, 'quote_count')
    print("Outliers in 'quote_count':")
    print(outliers_quote_count)
```

## Summarize Data

‚û°Ô∏è Assignment Tasks
- Create at least two charts to summarize and visualize the data
- Examples include: Number of posts per device; top 10 keywords with highest engagement

```python
# Group by language and count the number of posts
lang_counts = df.groupby('lang')['lang'].count()

# Sort lang_counts in ascending order
lang_counts = lang_counts.sort_values(ascending=True)

# Reverse the order for highest to lowest on x-axis
lang_counts = lang_counts.iloc[::-1]  # Reverse the order

# Create a bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(lang_counts.index, lang_counts.values)
ax.set_xlabel("Language")
ax.set_ylabel("Number of Posts")
ax.set_title("Number of Posts per Language (Descending Order)")  # Update title
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculate tweet lengths
df['tweet_length'] = df['text'].apply(len)

# Create the histogram
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.hist(df['tweet_length'], bins=20, edgecolor='black')  # Adjust bins as needed
plt.xlabel("Tweet Length (Characters)")
plt.ylabel("Frequency")
plt.title("Distribution of Tweet Lengths")
plt.show()
```

```python
# Time series plot to show the trend of tweet activity over time.
# Convert 'created_at' to datetime objects
df['created_at'] = pd.to_datetime(df['created_at'])

# Group by hour and count the number of tweets
hourly_counts = df.groupby(df['created_at'].dt.hour)['created_at'].count()

# Create the time series plot
plt.figure(figsize=(12, 6))
plt.plot(hourly_counts.index, hourly_counts.values)
plt.xlabel("Hour of Day")
plt.ylabel("Number of Tweets")
plt.title("Tweet Activity Over Time (Hourly)")
plt.xticks(range(24))  # Set x-axis ticks to represent hours
plt.grid(True)  # Add a grid for better readability
plt.tight_layout()
plt.show()
```
