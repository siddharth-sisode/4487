```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

## Getting data into Pandas

In this case we will load data from the statsmodels.org library

MTCARS is a traditional dataset used for learning.



```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
df['carmodel'] = df.index
print(df)
```

## Preview Data

Use the following data dictionary to understand the attributes available

|Attribute | Description |
|---------|--------------|
|	mpg	|Miles/(US) gallon|
|	cyl|	Number of cylinders|
|	disp|	Displacement (cu.in.)|
|hp|	Gross horsepower|
|drat|	Rear axle ratio|
|wt|	Weight (1000 lbs)|
|qsec|	1/4 mile time|
|vs|	Engine (0 = V-shaped, 1 = straight)|
|am|	Transmission (0 = automatic, 1 = manual)|


# Create basic charts

- Line chart
- Bar chart
- Histogram
- Box plot
- Scatter plot




```python
df.plot(x='carmodel', y='mpg', kind='line', title="MPG by Car")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Bar chart
df.plot(x='carmodel', y='mpg', kind='bar', title="MPG by Car", rot=45, figsize=(15, 6), width=0.3)
plt.tight_layout()
plt.show()
```

```python
#Histogram
df['mpg'].plot(kind='hist', bins=10, title="MPG Distribution")
```

```python
#Boxplot
df['mpg'].plot(kind='box', title = 'Miles Per Gallon')
```

```python
#Scatter plot
df.plot.scatter(x='mpg',y='hp',c='wt')
```

## Your Turn: Create an Additional Chart

 Assignment Tasks
- Create a pie chart showing the count of cars by transmission type
- Create a bar chart with different colors for each car model

```python
# Pie chart for transmission type

# Count the occurrences of each transmission type
transmission_counts = df['am'].value_counts()

# Create the pie chart
plt.figure(figsize=(6, 6))  # Adjust figure size if needed
plt.pie(transmission_counts, labels=transmission_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Count of Cars by Transmission Type\n(0 = Automatic Transmission, 1 = Manual Transmission)')
plt.show()


```

```python
# Bar chart with different colors for each car model
car_models = df['carmodel'].unique()
num_cars = len(car_models)

# Generate a color map using viridis and normalize it
cmap = plt.get_cmap('viridis')
norm = mpl.colors.Normalize(vmin=0, vmax=num_cars - 1)

# Create Bar Chart
plt.figure(figsize=(12, 6))
plt.bar(df['carmodel'], df['mpg'], color=[cmap(norm(i)) for i in range(num_cars)])
plt.xlabel('Car Model')
plt.ylabel('Miles Per Gallon (MPG)')
plt.title('MPG by Car Model with Different Colors')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
